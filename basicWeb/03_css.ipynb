# CSS

- HTML vs CSS:  
  \~>  정보, 구조화 vs styling의 정의
- 각자 문법이 다름(별개의 언어) ~ css는 html에 의존하긴 함

## 기본 사용법

- selector: 어떤 범주를
- 선언
  - property: 어떤 것을
  - value: 어떻게 적용   
  
          
- 활용하기
  - 인라인(라인마다 적용)
  - 임베딩(헤드에서 내부참조)
  - 링크(외부참조)  <\~ 일반적으로 이렇게 활용
  
## 선택자

- \* : 전체에 해당
- tag selector: tag명으로 적용
- class selector: .<class\> 으로 적용
- id selector: #<id\> 으로 적용  
  \~> id > class > tag 순으로 우선순위 적용 + css에서 나중 정의한 것 적용
    
- !important: 강조하기 위해서 이걸로 무조건 적용시키기 위해서

- group selector: ,로 구분하여 다수 적용 가능
- 인접형제: a + b + c \~> a가 나오고 b가 나오고 c가 나오면 적용
- 일반형제: a ~ b \~> a 뒤에 b가 나오면 다 적용
- 자식: parent > child \~> 바로 자식 노드에만 적용
- 후손: parent child \~> 부모노드 뒤 모든 자식에게 적용

## 프로퍼티

- 값의 단위
  - 키워드: ex) red, bold
  - 크기단위: ex) px, %, em(상대배수단위), rem, viewport
  - 색깔
  
## Box Model

- 구성
  - 마진: 외부여백(색지정x) \~> border바깥의 영역
  - border: 테두리영역(테두리)
  - 패딩: 내부여백(이미지 등은 여기까지 적용)
  - 내용: 실제내용위치
  
- display 속성:  
    css에서 display: xxx 로 변경가능
  - block
     - 항상 새로운 라인에서 시작
     - 전체영역(가로폭)을 차지(길막러) \~> width로 제어가능(margin은 유지)
     - in-line 요소 포함
     - ex) div, h, hr, ...
     
  - in-line
     - content너비만큼 가로폭 차지
     - 바로 뒤에 요소가 올 수 있음 \~> 중간 삽입 목적 tag들
     - width, height, margin 등 property x  
         \< margin-left/right // padding은 가능 \> \~> 위로 벌릴 수가 없음
     - ex) a, span, input, strong, button ... 
     
  - in-line block
     - 두 속성 합친 것: in-line이면서 margin등 가능
     
  - None
     - 요소를 사라지게(공간을 없앤다)  
         \vs hidden(visibility)
         
- background
  - image
  - size: contain, cover ...
  - repeat, attachment ...  
  \~> image url로 불러서 cover로 돌리는게 속편함
  
- font
  - family참조하여 다양한 구현 가능
  - 필수값을 알아두면 shorthand(한줄로) 표현가능 \~> 패밀리, 사이즈
  
## Position
- 기본적으로 static \~> 쌓여가는 구조
- 유형
  - 상대적위치:
    - positiob: relative + 위치치정  
    \~> 원래 있던 위치대비 위치 지정 **기억 포지션은 원위치!!**
    
  - 절대적 위치:
    - 기준점 필요 \~> parent중 fixed/static(지정안되면 body까지 참조올라감)
    
  - fixed:
    - 화면을 기준으로 고정된 위치
    
  - float:
    - 일정한 영역을 차지하는 그림
    - 글들이 피해가는 그림을 만들기

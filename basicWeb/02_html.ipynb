## Hyper Text Markup Language

- 순차적으로 진행되는 것이 아니라 다양한 페이지에 참조가능(일반 텍스트 > 페이지순으로 연결되어 있음)  
 \~> 규칙: TP(Text Protocol) >> HTTP(s) (HP끼리 통신하는 규칙을 의미)  
- Markup: 글에게 각 의미를 부여(ex) 문단나누기, heading)
- Language: 역할을 부여하기 위한 표시 언어(연산 등 x)

\~> HTML파일 == HTML로 작성된 문서파일(1)  

- CSS: ML 의미부여  >>  스타일부여

## HTML 기본구조

- doctype 선언부: 최상단에 존재하여 문서구분
- HTML 요소: 최상위 요소로 문서의 root를 의미
    - head: 문서제목, 문자코드(인코딩) 등  >>  브라우전에 노출
        - 탭에 써잇는 이름
        - CSS 포함 등도 표시
        - open graph도 여기에 해당
    - body: HTML(browser)에서 노출되는 실제 내용
    
## Tag와 DOM TREE(document of model)
   - tage들은 tree로 연결되어 있음(doc \~> html \~> head, body)
   
### 요소(element)
   - 소문자로 작성이 원칙
   - 요소간 중첩 가능  
    ** html은 오류를 내지 않음!! \~ 조심히 작성하여야 함 **
   - 속성도 부여 가능
   - self-closing element 존재  \~> meta, br  

### 속성(attribute)
   - 태그에는 속성 지정 가능(id, style, class)

### DOM TREE
   - tree: 부모와 자식 관계 // 형제 관계
   
### Sementic Tag
   - 의미 있는 정보의 그룹을 태그로 표현하여 의미를 가지는 태그들을 활용하기 위한 노력  
    ~\> 개발자/사용자 모두에게 의미가 있음  
   - <div\>: 공간을 정의(의미는 x)
   - header, footer 등으로 이름을 할당해주면 편함

### table
   - tr으로 각 row작성
   - 명시적으로 thead, tbody를 나누는 것 권장
   - caption으로 제목작성 가능
   - colspan, rowspan으로 셀병합가능
   
### form
   - 제출하는 행위 + 제출받아 가는 곳이 중요
   - action: 어디로 제출!
       - 지정하지 않으면 본 위치로 다시 보냄
   - tags:
       - input:
         - attribute: input이 가지고 있을 값(default)
         - type: text, number, password, email, date, radio(name="language"), select(option), color
         - submit: 제출해 주는 것 // value=그 이름  
          \~> 제출하면 url에 요청변수가 변하여 있는 것 확인 가능  
          \~> name을 지정해 주어야 입력받아서 가능  
         - placeholder: 입력창에 적용해서 뒷배경

# DataBase

- 여러사람이 공유하고 사용할 목적
- 통합관리되는 정보 집합

## 장점

- 중복최소화
- 무결성, 일관성, 독립성, 표준화
- 보안 유지

## 관계형 데이터베이서 관리 시스템(RDBMS)

- 관계를 표현하기 위해 table사용(like dataframe) \~> 식별자 존재(ID)


## 구성

### 스키마
- 자료의 구조, 표현방법, 관계 등을 정의한 구조
    - ex) id - int // age - int // email - text
    
### 테이블
- 정의된 요소들의 집합
- col \~ row로 정의
- pk(primary key): 식별자(id)

# SQL

- RDBMS의 관리를 위해 설계된 프로그래밍언어

## 문법

- 데이터정의언어(DDL): 테이블, 스키마 정의하기 위한 명령
    - ex) create, drop, alter
- 데이터조작언어(DML): 저장, 수정, 삭제, 조회 위한 언어(CRUD)
    - ex) insert, update, delete, update
- 데이터제어언어(DCL): DB 사용제어 목적

### CRUD

#### INSERT
- INSERT INTO fname (col1, col2 ...) <- col_names  
    VALUES (val1, val2 ...) <- target_value
    - 지정안하면 빈셀로 저장
    - 빈 값 없으면 col 생략가능
    
- PRIMARY KEY를 지정하지 않으면 PK옵션을 가진 'rowid'를 생성

- table 설정시 스키마에 NOT NULL 설정하여 빈 값허용 안하기 가능  
    \~> 

#### SELECT
- SELECT _colnames_ FROM fname
    - 이어서 LIMIT _num_으로 개수 제한 가능
    - LIMIT _num_ OFFSET _num_: OFFSET만큼 넘어가서 LIMIT갯수 만큼 가져오기
    - WHERE col_name=value: 해당 col 조건에 맞는 것만 선택
- SELECT DINSTINCT col_name FROM fname
    - 중복 제외하고 해당 col 값 select
    
#### DELETE
- DELETE FROM table  
    WHERE rowid
    - 보통 중복된 값이 없는 pk(rowid)값을 사용하여 삭제
    
- 삭제하면 해당 행을 채우는 것을 우선으로 처리(빈 값을 채움)
    - 빈 값을 재사용하지 않게 하기 위해서 schema지정 시 id --- AUTOINCREMENT옵션 추가
    - 옵션 사용 않는 것이 권장사항 \~ 자기 옵션에 따라 설정할 것!!

#### UPDATE
- UPDATE fname  
    SET col_name='target'  
    WHERE index=xxx
    
### 부가 기능

- 조건 부로 호출 하기:
    - SELECT col_name FROM fname WHERE condition;
        - condition은 and/ 으로 연속 조건 가능
        
- COUNT:
    - SELECT COUNT(col_name) FROM fname WHERE condition;
    
- AVG, SUM, MIN, MAX
    - INTEGER일때만 연산 가능
    - WHERE 조건 추가 가능

- LIKE(wild cards): 패턴을 확인하여 해당값 반환
    - pattern:
        - _: 반드시 이 자리에 한 개 문자 존재
        - %: 이자리에 문자열이 있을 수도 없을 수도
    - SELECT * FROM fname WHERE _colname_ LIKE '2%_%_'
    
- ORDER(정렬)
    - SELECT col_name FROM fname ORDER BY col1(...) ASC(or DESC)
        - LIMIT는 ORDER BY 뒤에 해야 됨
        
- table 이름 변경
    - ALTER TABLE fname RENAME to new_fname
    
-  새로운 col 추가
    - ALTER TABLE fname ADD COLUMN new_col 
        - NOT NULL 조건시 문제발생
            - NOT NULL 제거
            - default값 추가로 해결 가능
